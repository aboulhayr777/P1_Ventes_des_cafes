### Dataset description : 
The Dirty Cafe Sales dataset contains 10,000 rows of synthetic data representing sales transactions in a cafe. This dataset is intentionally "dirty," with missing values, inconsistent data, and errors introduced to provide a realistic scenario for data cleaning and exploratory data analysis (EDA). It can be used to practice cleaning techniques, data wrangling, and feature engineering.

### Columns (8 columns)
**Transaction ID** : A unique identifier for each transaction.,  
**Item** : The name of the item purchased.,  
**Quantity**: The quantity of the item purchased.,  
**Price Per Unit** : The price of a single unit of the item.,  
**Total Spent** : The total amount spent on the transaction.,  
**Payment Method** : The method of payment used.,   
**Location** : The location where the transaction occurred.,   
**Transaction Date** : The date of the transaction.,

```python
# Pour les colonnes "Quantity", "Price Per Unit" et "Total Spent"

# Conversion des colonnes
cols = ["Quantity", "Price Per Unit", "Total Spent"]
for col in cols : 
    coffee[col] = pd.to_numeric(coffee[col], errors="coerce")
```

```python
for col in cols :
    print(coffee[col].dtype)
```

```python
coffee["Total Spent"].unique()
```

```python
# Gestion des valeurs manquantes pour les 3 colonnes
```

```python
coffee[cols].sample(45)
```

```python
mask = coffee[cols].isna().sum(axis=1) >= 2

coffee[mask].shape[0]
```

```python
# Création des masks
mask_total = (
    coffee["Total Spent"].isna() &
    coffee["Quantity"].notna() &
    coffee["Price Per Unit"].notna()
)

mask_quantity = (
    coffee["Quantity"].isna() &
    coffee["Price Per Unit"].notna() &
    coffee["Total Spent"].notna() &
    (coffee["Price Per Unit"] != 0)
)

mask_price_per_unit = (
    coffee["Price Per Unit"].isna() &
    coffee["Quantity"].notna() &
    coffee["Total Spent"].notna() &
    (coffee["Quantity"] != 0)
)
```

```python
coffee[mask_price_per_unit][cols]
```

```python
# Imputation des valeurs manquantes
    # Pour Total Spent
coffee.loc[mask_total, "Total Spent"] = coffee.loc[mask_total, "Quantity"] * coffee.loc[mask_total, "Price Per Unit"]
    
    # Pour Quantity
coffee.loc[mask_quantity, "Quantity"] = coffee.loc[mask_quantity, "Total Spent"] / coffee.loc[mask_quantity, "Price Per Unit"]

    # Pour Price Per Unit
coffee.loc[mask_price_per_unit, "Price Per Unit"] = coffee.loc[mask_price_per_unit, "Total Spent"] / coffee.loc[mask_price_per_unit, "Quantity"]
```

```python
coffee[mask].shape[0]
```

```python
(58/10_000)*100
```

```python
# Exclusion des 58 lignes dans lesquelles 2 colonnes parmi les variables financières ne sont pas renseignées
coffee_clean = coffee[coffee[cols].isna().sum(axis=1) == 0]
coffee_clean
```

Les variables financières (Quantity, Price Per Unit, Total Spent) ont été reconstruites uniquement lorsque une seule des trois composantes était manquante, en utilisant la relation arithmétique entre elles. Les cas incohérents ou insuffisamment renseignés ont été exclus.

Une faible proportion des données (0,58%) présentait plusieurs valeurs manquantes simultanément sur les variables clés (quantité, prix unitaire, total). Ces observations ont été exclues afin de garantir la cohérence des analyses, sans impact significatif sur les résultats globaux.

```python
coffee_clean.to_csv("cleaned_cafe_sales.csv", index=False)
```
